```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sklearn.datasets
```

```python
d1 = np.loadtxt("example_1.txt")
d2 = np.loadtxt("example_2.txt")
print(d1.shape, d2.shape)
```

```python
df = pd.DataFrame(
    {
       "value": np.concatenate((d1, d2)),
       "type": np.concatenate((np.ones(d1.shape), np.zeros(d2.shape))) 
    }    
)
```

```python
plt.figure(figsize = (12, 6))
sns.boxplot(x = df['value'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'type', y = 'value', data = df)
sns.swarmplot(x = 'type', y = 'value', data = df, color = '#000', alpha = 0.35)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'type', y = 'value', data = df, whis = 0.2)
sns.swarmplot(x = 'type', y = 'value', data = df, color = '#000', alpha = 0.35)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.violinplot(x = 'type', y = 'value', data = df)
sns.swarmplot(x = 'type', y = 'value', data = df, color = '#000', alpha = 0.35)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.violinplot(x = 'type', y = 'value', data = df, inner = 'quartile')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.violinplot(x = 'type', y = 'value', data = df, inner = 'quartile', bw = 0.1)
plt.show()
```

```python
data = sklearn.datasets.load_iris()
```

```python
iris = pd.DataFrame(data = data['data'], columns = data['feature_names'])
iris['target'] = data['target']
iris['spec'] = iris['target'].apply(lambda t : data['target_names'][int(t)])
iris.head()
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(x = 'spec', y = 'sepal length (cm)', data = iris)
sns.swarmplot(x = 'spec', y = 'sepal length (cm)', data = iris, color = 'black', alpha = 0.3)
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.violinplot(x = 'spec', y = 'sepal length (cm)', data = iris, inner = 'quartile')
sns.swarmplot(x = 'spec', y = 'sepal length (cm)', data = iris, color = 'black', alpha = 0.3)
plt.show()
```
