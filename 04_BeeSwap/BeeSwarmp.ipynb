```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d1 = np.loadtxt("example_1.txt")
d2 = np.loadtxt("example_2.txt")
print(d1.shape, d2.shape)
```

```python
df = pd.DataFrame(
    {
       "value": np.concatenate((d1, d2)),
       "type": np.concatenate((np.ones(d1.shape), np.zeros(d2.shape))) 
    }    
)
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize = (12, 6))
sns.swarmplot(x = df['value'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.swarmplot(x = 'type', y = 'value', data = df)
plt.show()
```

```python
import sklearn.datasets
```

```python
data = sklearn.datasets.load_iris()
```

```python
data
```

```python
iris = pd.DataFrame(data = data['data'], columns = data['feature_names'])
iris['target'] = data['target']
iris['spec'] = iris['target'].apply(lambda t : data['target_names'][int(t)])
iris.head()
```

```python
plt.figure(figsize = (12, 8))
sns.swarmplot(x = 'spec', y = 'sepal length (cm)', data = iris)
# sns.swarmplot(x = 'spec', y = 'petal length (cm)', data = iris)
plt.show()
```
